```python
data_subsidiosCD=pd.read_csv(r'C:/Users/Miguel/Desktop/Ciencia de Datos/TP/Compilado/TP_clusteria-master/aportes_colectivoscd.csv', delimiter=';')
print(data_subsidiosCD.shape)
print("Dataset Colectivos de Corta Distancia. Están mal asignados los nombres de las columnas :S")
data_subsidiosCD.tail()
```

```python
data_subsidiosCD_CABA['mes'] = data_subsidiosCD_CABA['mes'].apply(lambda x: x.replace('/', '-') if '/' in str(x) else x)
```

```python
data_subsidiosCD_CABA.columns = ['periodo','cuit','provincia','municipio','monto','tipo','año','mes']
print(data_subsidiosCD_CABA.shape)
data_subsidiosCD_CABA.tail(15)
```

```python
# cambie monto por tipo en este bloque de codigo

subisidiosCD_tipo = pd.pivot_table(data_subsidiosCD_CABA,values = 'monto',columns = 'tipo', index = data_subsidiosCD_CABA.mes.values, aggfunc = np.sum)
subisidiosCD_tipo.head(3)
```

```python
# eliminar datos 2017
# agrupar por año y por tipo de subsidio
# agragrle al cantidad total de pasajeros trasladados para 2018 ¿Las empresas que mas subsidios cobran, trasladan mas pasajeros?
# ¿Queremos eliminar algun componente? 
# Clustering por lineas, barrios recorridos, distancias recorridas
# En IGJ puedo saber quienes son los accionistas de las empresas?
```

```python
# genero un nuevo df para 2018

results3_2018 = results3.loc[results3['año'] == 2018]
```

```python
np.shape(results3_2018)
```

```python
# vemos que tiene nans en razon social. Es decir, que el cuit no se encontro en la base del sistau cuando hicimos el merge

col_NaN = results3_2018.isnull().any()
print(col_NaN,'\r\n')
cant_NaN = results3_2018.isnull().sum()
print(cant_NaN,'\r\n')
```

```python
# confirmo con un unique #

results3_2018['razon_social'].unique()
```

```python
#quiero ver si las empresas para las cuales no se encontró la razón social en el merge tienen monto#

# primero reemplazo los NaN por un texto simple para poder filtrarlo 
# ESTO PODEMOS REVISARLO YA QUE EXISTE UNA FORMA DE FILTRAR POR NANS SIN NECESIDAD DE REEMPLAZARLOS

results4_2018 = results3_2018.fillna('SOY_NAN')
print(results4_2018.shape)
results4_2018.head(3)
```

```python
# Agrupo y veo que existen montos subsidiados para empresas que no tenemos la razón social, todos por combustible

results4_2018[results4_2018['razon_social'] == 'SOY_NAN'].groupby('tipo')['monto'].sum().reset_index()
```

```python
# Todos se otorgaron desde el mes de agosto
# La explicación puede ser que la base de SISTAU que utilizamos tiene datos sólo hasta el mes de mayo18

results4_2018[results4_2018['razon_social'] == 'SOY_NAN'].groupby('mes')['monto'].sum().reset_index()
```

```python
#empresas con mas cantidad de subsidios - Los SOY_NAN estan casi al fondo

ranking2 = results4_2018.groupby(["razon_social"])["monto"].sum().sort_values(ascending=False)
ranking2.to_frame()
```

```python
#tabla pivote con el subsidio diferenciado entre comp gasoil y comp tarifaria

pivote2018 = pd.pivot_table(results4_2018,values = 'monto',columns = 'tipo', index = results4_2018.razon_social.values, aggfunc = np.sum)
pivote2018.head(3)
```
